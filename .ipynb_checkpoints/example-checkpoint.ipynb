```python
%%cython

cimport cython
cimport numpy as np
import numpy as np

@cython.boundscheck(False)
@cython.wraparound(False)
def _get_tao_cython(np.ndarray[float, ndim=2] X, 
                    np.ndarray[long long] ind,
                    np.ndarray[long long] reduced):
    cdef Py_ssize_t i,j,l
    cdef Py_ssize_t k = 0
    cdef Py_ssize_t N = X.shape[0]
    cdef Py_ssize_t P = X.shape[1]
    cdef Py_ssize_t M = len(reduced) # shape of tao
    cdef np.ndarray[float] tao = np.zeros(dtype=np.float32, shape=(M))
    cdef int start_id
    cdef int target_id, compared_id
    cdef int top_density_id
    cdef float _min, tmp
    
    startlist = np.argsort(ind)[np.searchsorted(np.sort(ind), reduced)]
    
    for i in range(M):
        target_id = reduced[i]
        start_id  = startlist[i] + 1
        _min = 1000.0
        if start_id < N:
            # now let's find the minimum distance from target_id to any higher density compaired_ids
            # compared_id = [ind[start_id], ind[start_id+1], ... ind[start_id+j],... ind[-1]]
            for j in range(N-start_id):
                compared_id = ind[j+start_id]
                tmp = 0.0
                for l in range(P):
                    tmp += (X[target_id, l]-X[compared_id, l])*(X[target_id, l]-X[compared_id, l])
                if tmp < _min:
                    _min = tmp
            tao[i] = _min
        else:
            top_density_id = i
    
    tao[top_density_id] = max(tao)
    return tao
```

```python
@interact(k=(5,1000))
def view_compress(k=15):
#     k = X.shape[0]/N
    global rho, reduced
    rho,reduced = mc.compress(X, k)
    mc.view_compressed(X, reduced)
```

```python
ind = np.argsort(rho)
tao = _get_tao_cython(X, ind, reduced)
```

```python
def get_thr_tao(tao, segment_value=5):
    v = np.sort(tao)
    critical_id = np.where(np.diff(v) > (max(tao)-min(tao))/segment_value)[0][0]
    thr = v[critical_id]
    return thr
```

```python
plot(tao[np.argsort(rho[reduced])],'o')
thr = get_thr_tao(tao, segment_value=5)
axhline(thr, ls='-.', c='c')
center_id   = reduced[np.where(tao>thr)[0]]
```
